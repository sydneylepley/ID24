```python
import pandas as pd 
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
diabetes = pd.read_csv('Diabetes.csv')
diabetes
```

```python
sns.scatterplot(x=diabetes['BMI'], y=diabetes['Glucose'])
plt.show()
```

```python
sns.relplot(data=diabetes, x= "BMI", y= "Glucose") 
```

```python
sns.relplot(data=diabetes, x="BMI", y="Glucose", hue="Pregnancies")
```

```python
sns.relplot(data=diabetes, x="BMI", y="Glucose", hue="Pregnancies", size="Pregnancies")
```

```python
sns.relplot(data=diabetes, x="BMI", y="Glucose", hue="Pregnancies", size="Pregnancies", sizes= (10,50))
```

```python
sns.relplot(data=diabetes, x="BMI", y="Glucose", hue="Pregnancies", size="Pregnancies", sizes= (10,50))
plt.xlim(18,37)
plt.ylim(50,140)

plt.show()
```
